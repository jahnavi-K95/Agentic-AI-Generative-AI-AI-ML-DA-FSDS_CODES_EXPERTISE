#  ARRAY CREATION FUNCTIONS

```python
import numpy as np
```

```python
#Create an array from a list
a = np.array([1,2,3])
print(" Array a:", a)
```

```python
#create an array with evenly spaced values
b = np.arange(0, 10, 2)
print("Array b:", b)
```

```python
b = np.arange(0, 10, 3)
print("Array b:", b)
```

```python
# Create an array filled with zeros

d = np.zeros((2,3), dtype=int)
print("Array d:\n", d)
```

```python
d = np.zeros((2,3))
print("Array d:\n", d)
```

```python
# Create an array filled with ones

d = np.ones((2,3), dtype=int)
print("Array d:\n", d)
```

```python
d = np.ones((3,2))
print("Array d:\n", d)
```

```python
#create an Identity matrix
f = np.eye(4)
print("Identity matrix f:\n", f)
```

# ARRAY MANIPULATION FUNCTIONS

```python
#Reshape an array

a1 = np.array([1,2,3])
reshaped = np.reshape(a1, (1,3)) #Reshape to 1*3
print("reshaped array:", reshaped)
```

```python
#Reshape an array

a1 = np.array([1,2,3])
reshaped = np.reshape(a1, (3,1)) #Reshape to 1*3
print("reshaped array:", reshaped)
```

```python
# Flatten an array
f1 = np.array([[1,2],[3,4]])
flattened = np.ravel(f1)
print("Flattened array:", flattened)
```

```python
# Flatten an array
f1 = np.array([[1,2],[3,4],[5,6],[7,8]])
flattened = np.ravel(f1) #Flatten 1D array
print("Flattened array:", flattened)
```

```python
# Transpose an array
e1 = np.array([[1,2],[3,4]])
transposed = np.transpose(e1)
print("Transposed array:\n", transposed)
```

```python
#Stacks arrays vertically
a2 = np.array([1,2])
b2 = np.array([3,4])
stacked = np.vstack([a2,b2])
print("stacked arrays:\n", stacked)
```
